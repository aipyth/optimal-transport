```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# C1 = lambda x, y: np.abs(x - y)
# C2 = lambda x, z: np.abs(x - z)
# C3 = lambda y, z: np.abs(y - z)

C1 = lambda x, y: np.power(x - y, 2)
C2 = lambda x, z: np.power(x - z, 2)
C3 = lambda y, z: np.power(y - z, 2)

C = lambda x, y, z: C1(x, y) + C2(x, z) + C3(y, z)

n = 100

X = np.linspace(0, 1, n)
Y = np.linspace(0, 1, n)
Z = np.linspace(0, 1, n)
```

```python
XV, YV = np.meshgrid(X, Y)

plt.matshow(C1(XV, YV))
```

```python
XVV, YVV, ZVV = np.meshgrid(X, Y, Z)

plt.matshow(C(XVV, YVV, ZVV)[99])
```
